In the following graph, we can see which letters (we don't care if they are consonants or vowels) appear the most in the chosen poems. The poems in the centre are the ones that appear the most. 

```python
%%dot -f svg
digraph e {
    rankdir=LR
    letter1 [label="The letter that appears the most" shape=box style=filled fillcolor=red]
    letter2 [label="The letter that appears the second-most" shape=box style=radial fillcolor=turquoise]
    letter3 [label="The letter that appears the third-most" shape=box style=filled fillcolor=lawngreen]
    {rank=same; letter1, letter2, letter3}
    {rank=same; e, n, l, r, i, s}
    {rank=same; Tanglerhuette, Osten, Abschied, Unterstand, Schneeglöckchen, Grau }
        e [label=e shape=circle style=filled fillcolor=red fontcolor=black]
        Tanglerhuette [label="Tangerhuette" sape=box]
        Osten [label="Fahrt zum Osten" shape=box]
        Abschied [label="Ungewisser Abschied" shape=box]
        Unterstand [label="Unterstand in Nacht" shape=box]
        Schneeglöckchen [label="Erste Schneegloeckchen 1915" shape=box]
        Grau [label=Grau shape=box]
        n [label=n shape=circle style=radial fillcolor=turquoise]
        l [label=l shape=circle style=radial fillcolor=lawngreen]
        r [label=r shape=circle style=radial fillcolor=lawngreen]
        i [label=i shape=circle style=radial fillcolor=lawngreen]
        s [label=s shape=circle style=radial fillcolor=lawngreen]
            letter2 -> Tanglerhuette [dir=none color=white]
            letter2 -> Osten [dir=none color=white]
            letter2 -> Abschied [dir=none color=white]
            letter2 -> Unterstand [dir=none color=white]
            letter2 -> Schneeglöckchen [dir=none color=white]
            letter2 -> Grau [dir=none color=white]
            Tanglerhuette -> e [color=red]
            Osten -> e [color=red]
            Abschied -> e [color=red]
            Unterstand -> e [color=red]
            Schneeglöckchen -> e [color=red]
            Grau -> e [color=red]
            Tanglerhuette -> n [color=turquoise]
            Osten -> n [color=turquoise]
            Abschied -> n [color=turquoise]
            Unterstand -> n [color=turquoise]
            Schneeglöckchen -> n [color=turquoise]
            Grau -> n [color=turquoise]
            Tanglerhuette -> i [color=lawngreen]
            Osten -> l [color=lawngreen]
            Abschied -> r [color=lawngreen]
            Unterstand -> s [color=lawngreen]
            Schneeglöckchen -> r [color=lawngreen]
            Grau -> l [color=lawngreen]
    }
```

In the next graph, we can also see which letters appear the most in the poems. However, this time, we are distinguishing between consonants and vowels. Vowels are coloured red, consonants blue. Furthermore, the letters that are placed in the centre of the graph are the ones that are the most connected, namely, the ones that appear the most.

It stands out that concerning the consonants, the ones that are used the most (r, n, l) are sonorants, so there are not many "hard" consonants like voiceless stops or affricates.
